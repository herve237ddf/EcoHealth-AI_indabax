```python
def correction_recuperation_contenu_colonne(df, colone):
    dfe= df.copy()
    dfe[colone] = pd.to_datetime(dfe[colone], errors='coerce')
    dfe= dfe.dropna()
    dfe[colone]= dfe[colone].dt.day + (dfe[colone].dt.month/100)

    dfn= df.copy()
    dfn[colone] = pd.to_numeric(dfn[colone], errors='coerce')
    dfn= dfn.dropna()

    df_fusion = pd.concat([dfe,dfn],axis=0, ignore_index=True, sort=False)

    return df_fusion
```

```python
dft = df.copy()
for ele in lis_colonne_pb:
    dft = correction_recuperation_contenu_colonne(dft, ele)
```

# Feature engineering

```python
dft.drop(columns=['month', 'year'], inplace=True)
```

```python
dft['quarter']      = dft['time'].dt.quarter
dft['day_of_year']  = dft['time'].dt.dayofyear
dft['month']  = dft['time'].dt.month
dft['month_sin']    = np.sin(2 * np.pi * dft['month'] / 12)   # encodage cyclique
dft['month_cos']    = np.cos(2 * np.pi * dft['month'] / 12)
dft['is_dry_season']= dft['month'].isin([11,12,1,2,3]).astype(int)

# ── Variables dérivées (indicateurs de pollution potentielle) ─────────────────
dft['temp_amplitude']   = dft['temperature_2m_max'] - dft['temperature_2m_min']
dft['sunshine_ratio']   = dft['sunshine_duration'] / (dft['daylight_duration'] + 1e-6)
dft['is_no_wind']       = (dft['wind_speed_10m_max'] < 5).astype(int)  # stagnation
dft['is_no_rain']       = (dft['precipitation_sum'] < 0.1).astype(int)  # pas de lessivage
```

```python
# On s'assure que le dataset est en "étages" (Villes puis Dates)
dft = dft.sort_values(['city', 'time']).reset_index(drop=True)

# 1. Créer une colonne temporaire pour identifier si le jour est sec (Seuil 0.1mm)
dft['is_dry'] = (dft['precipitation_sum'] < 0.1).astype(int)

# 2. Ta boucle existante avec l'ajout du compteur de jours secs
for lag in [1, 2, 3, 4, 5, 6, 7]:
    # Tes lags existants
    dft[f'temp_lag{lag}'] = dft.groupby('city')['temperature_2m_mean'].shift(lag)
    dft[f'wind_lag{lag}'] = dft.groupby('city')['wind_speed_10m_max'].shift(lag)
    dft[f'preci_lag{lag}'] = dft.groupby('city')['precipitation_sum'].shift(lag)
    dft[f'evapo_lag{lag}'] = dft.groupby('city')['et0_fao_evapotranspiration'].shift(lag)

# 3. Ajout du compteur de jours sans pluie (consécutifs)
# On utilise cumsum sur l'inverse de 'is_dry' pour créer des blocs de sécheresse
dft['dry_group'] = (dft['is_dry'] == 0).groupby(df['city']).cumsum()
dft['consecutive_dry_days'] = dft.groupby(['city', 'dry_group']).cumcount()

# 4. On décale de 1 pour ne pas compter le jour actuel (comme tu le souhaitais)
dft['consecutive_dry_days'] = dft.groupby('city')['consecutive_dry_days'].shift(1).fillna(0).astype(int)

# Nettoyage des colonnes temporaires
dft.drop(columns=['is_dry', 'dry_group'], inplace=True)
```

```python
dft['temp_roll7'] = dft.groupby('city')['temperature_2m_mean'].transform(
    lambda x: x.rolling(7, min_periods=1).mean()) # etude de la moyenne des temperature durant les 7 derniers jours
```

```python
dft["indice_stagnation"] = (1/(dft["wind_speed_10m_max"]+0.1)) * (1- (1-dft["is_no_rain"])*0.5)
```

```python
#etude des anomalie pour chacune des colonnes suivantes durant les 7 derniers jours (les lags derniers jours)
for lag in [1, 2, 3, 4, 5, 6, 7]:
    # Tes lags existants
    dft[f'anomalie_temp_lag{lag}'] = dft["temperature_2m_mean"] - dft[f'temp_lag{lag}']
    dft[f'anomalie_wind_lag{lag}'] = dft["wind_speed_10m_max"] - dft[f'wind_lag{lag}']
    dft[f'anomalie_preci_lag{lag}'] = dft["precipitation_sum"] - dft[f'preci_lag{lag}']
    dft[f'anomalie_evapo_lag{lag}'] = dft["et0_fao_evapotranspiration"] - dft[f'evapo_lag{lag}']
```
